**IMAGE TO PENCIL SKECTH WITH PYHTON**

```python
# mount google drive in google colab
from google.colab import drive
drive.mount('/content/drive')
```

```python
# import libraries including OpenCV
import numpy as np
import pandas as pd
from google.colab.patches import cv2_imshow
import cv2
from PIL import Image
```

```python
# read lion image from google drive
lion_image = cv2.imread('/content/drive/MyDrive/Image/lion.jpg')
cv2_imshow(lion_image)
```

```python
# convert input image into grayscale image
gray_image = cv2.cvtColor (lion_image, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray_image)
```

```python
# inert pixel values of gray image
invert_image = cv2.bitwise_not(gray_image)
cv2_imshow(invert_image)
```

```python
# blur the image using Gaussian Blur
blur_image = cv2.GaussianBlur(invert_image, (15,15), 0)
cv2_imshow(blur_image)
```

```python
# invert the pixel values of blurred image
blurinvert_image = cv2.bitwise_not(blur_image)
cv2_imshow(blurinvert_image)
```

```python
# divide the grayscale image by the inverted blurred image using OpenCV
pencil_sketch = cv2.divide(gray_image, blurinvert_image, scale = 255 -0 )
cv2_imshow(pencil_sketch)
cv2.imwrite ("pencil_sketch.png" , pencil_sketch)
```

**Finally got the pencil sketch of original lion image.**
